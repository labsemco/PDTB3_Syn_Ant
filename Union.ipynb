## Método para recuperar los sinónimos de una palabra por tipo

```python
def synonyms(wn, label, word : str):
    syn = list() # word's synonyms
    ant = list() # word's antonyms
    
    for synset in wn.synsets(word, pos=label):
        for lemma in synset.lemmas():
            #print(synset, lemma)
            if lemma.name() not in syn:
                print(lemma.name())
                syn.append(lemma.name())
            if lemma.antonyms():
                if lemma.antonyms()[0].name() not in ant:
                    ant.append(lemma.antonyms()[0].name())
    return syn,ant
```

```python
df = pd.read_csv(carpeta + 'sets.csv', index_col=0)
```

```python
df.head()
```

```python
ant_sin = {'max':530}
for index, row in df.iterrows():
    sin = row['Sinonimos'].split(',')
    ant = row['Antonimos'].split(',')
    #print('sin={} and ant={}'.format(sin, ant))
    ant_sin[int(index)] = {'synonyms':sin, 'antonyms':ant}
```

# SETS CREADOS MANUALMENTE A PARTIR DE LOS CONJUNTOS SINONIMOS Y ANOTONIMOS (ELISEO)

```python
sets = pd.read_csv('insumos/' + 'Conjuntos_V18mayo2022.csv')
```

```python
sets.iloc[:1]
```

```python
ant_syn = {'max': max(sets['Index'])}
for index, row in sets.iterrows():
    a=row[2].replace("[","").replace("]","")
    b=row[3].replace("[","").replace("]","")
    c=a.split(",")
    d=b.split(",")
    syn1=set()
    ant1=set()
    for p in c:
      #print(p.replace("'","").replace(" ",""))
      syn1.add(p.replace("'","").replace(" ",""))
    for p in d:
      #print(p.replace("'","").replace(" ",""))
      ant1.add(p.replace("'","").replace(" ",""))
    syn = list(syn1)
    ant = list(ant1)
    ant_syn[row['Index']] = {'synonyms': syn, 'antonyms': ant}
ant_syn
```

```python
limits = {'ADJ': [1,232],'NOUN': [232,232+119], 'VERB': [232+119,232+119+145],'ADV':[232+119+145,232+119+145+35]}

with open('salida/' 'antonymy_synonymy_Sets3.pickle', 'wb') as pickle_file:
        pickle.dump((ant_syn, limits), pickle_file)
```
