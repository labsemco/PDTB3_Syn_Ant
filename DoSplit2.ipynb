```python
import pandas as pd
import numpy as np
import pickle
```

# Load full representation for enriched sets

```python
fold = './Contrast_Concession/filesPickle/'

infile = open(fold + 'PDTB-Comparison-contrast-concession-explicit-implicit-RepresentationsCHANGED.pickle','rb')
data = pd.DataFrame(pickle.load(infile))
infile.close()

name = 'salida/' +  'antonymy_synonymy_Sets3.pickle'
infile = open(name, 'rb')
ant_syn, limits= pickle.load(infile)
```

# Do a respective split

```python
dataComb = {'Main index':[], 'Category':[], 'Relation Type':[], 'Arg1 ant_syn':[], 'Connective':[], 'Arg2 ant_syn':[],'Type':[]}

for index, row in data.iterrows():
    mIndex = row['Main index']
    cat = row['Category']
    relType = row['Relation Type']
    conn = row['Connective']
    typess = row['Type']
    
    arg1ADJ = row['Arg1 ant_syn'][:limits['ADJ'][1]]
    arg1NOUN = row['Arg1 ant_syn'][limits['NOUN'][0]:limits['NOUN'][1]]
    arg1VERB = row['Arg1 ant_syn'][limits['VERB'][0]:limits['VERB'][1]]
    arg1ADV = row['Arg1 ant_syn'][limits['ADV'][0]:limits['ADV'][1]]
    
    arg2ADJ = row['Arg2 ant_syn'][:limits['ADJ'][1]]
    arg2NOUN = row['Arg2 ant_syn'][limits['NOUN'][0]:limits['NOUN'][1]]
    arg2VERB = row['Arg2 ant_syn'][limits['VERB'][0]:limits['VERB'][1]]
    arg2ADV = row['Arg2 ant_syn'][limits['ADV'][0]:limits['ADV'][1]]
    
    # Choose the class
    #Para poder obtener los trains y los pickles por clase de palabras
    sub = arg1ADJ + arg1NOUN + arg1VERB + arg1ADV
    obj = arg2ADJ + arg2NOUN + arg2VERB + arg2ADV
    
    if np.sum(sub)!=0 and np.sum(obj)!=0:
        dataComb['Main index'].append(mIndex)
        dataComb['Category'].append(cat)
        dataComb['Relation Type'].append(relType)
        dataComb['Arg1 ant_syn'].append(sub)
        dataComb['Connective'].append(conn)
        dataComb['Arg2 ant_syn'].append(obj)
        dataComb['Type'].append(typess)
dataComb = pd.DataFrame(dataComb)
```
